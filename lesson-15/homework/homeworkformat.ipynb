```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Task 1: Basic Plotting
x = np.linspace(-10, 10, 400)
y = x**2 - 4*x + 4

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='$f(x) = x^2 - 4x + 4$', color='blue')
plt.title('Quadratic Function Plot')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Task 2: Sine and Cosine Plot
x = np.linspace(0, 2 * np.pi, 400)
sin_y = np.sin(x)
cos_y = np.cos(x)

plt.figure(figsize=(8, 5))
plt.plot(x, sin_y, label='sin(x)', color='red', linestyle='-', marker='o')
plt.plot(x, cos_y, label='cos(x)', color='green', linestyle='--', marker='x')
plt.title('Sine and Cosine Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Task 3: Subplots - 2x2 Grid
x1 = np.linspace(-2, 2, 400)
x2 = np.linspace(0, 2 * np.pi, 400)
x3 = np.linspace(0, 2, 400)
x4 = np.linspace(0, 10, 400)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Top-left: Cubic Function
axs[0, 0].plot(x1, x1**3, color='purple')
axs[0, 0].set_title('Cubic Function')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('f(x)')
axs[0, 0].grid(True)

# Top-right: Sine Function
axs[0, 1].plot(x2, np.sin(x2), color='blue')
axs[0, 1].set_title('Sine Function')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('sin(x)')
axs[0, 1].grid(True)

# Bottom-left: Exponential Function
axs[1, 0].plot(x3, np.exp(x3), color='green')
axs[1, 0].set_title('Exponential Function')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('e^x')
axs[1, 0].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Task 4: Scatter Plot
np.random.seed(0)
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)

plt.figure(figsize=(8, 5))
plt.scatter(x, y, c='magenta', marker='^')
plt.title('Random Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)
plt.show()
```

```python
# Task 5: Histogram
data = np.random.normal(0, 1, 1000)  # Generating random normal data

plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', alpha=0.7)
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Task 6: 3D Plotting (Improved)
x = np.linspace(-5, 5, 50)  # Reduced from 100 to 50
y = np.linspace(-5, 5, 50)  # Reduced from 100 to 50
x, y = np.meshgrid(x, y)
z = np.cos(x**2 + y**2)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('3D Surface Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')
fig.colorbar(surface, shrink=0.5, aspect=5)
plt.show()
```

```python
# Task 7: Bar Chart
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [200, 150, 250, 175, 225]

plt.figure(figsize=(8, 5))
plt.bar(products, sales, color=['red', 'green', 'blue', 'orange', 'purple'])
plt.title('Sales Data')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.grid(True, axis='y')
plt.show()
```

```python
# Task 8: Stacked Bar Chart
categories = ['Category A', 'Category B', 'Category C']
time_periods = ['T1', 'T2', 'T3', 'T4']
data = np.array([[5, 3, 4, 7], [2, 4, 6, 8], [3, 5, 2, 6]])

fig, ax = plt.subplots(figsize=(8, 5))
bottom = np.zeros(4)

# Plot each category stacked on top of the previous
for i, category in enumerate(categories):
    ax.bar(time_periods, data[i], bottom=bottom, label=category)
    bottom += data[i]

ax.set_title('Stacked Bar Chart')
ax.set_xlabel('Time Periods')
ax.set_ylabel('Values')
ax.legend()
plt.grid(True, axis='y')
plt.show()
```
